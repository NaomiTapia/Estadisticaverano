Introducción
El presente proyecto tiene como objetivo analizar los factores que influyen en los costos de los seguros médicos en un conjunto de datos obtenido de una fuente pública. Estos datos incluyen variables como edad, sexo, índice de masa corporal (BMI), número de hijos, hábito de fumar, región geográfica y los cargos médicos (charges). A través de un enfoque estadístico, se busca determinar si existe una relación significativa entre estas variables independientes y el monto que las aseguradoras cobran a los clientes. Este análisis resulta crucial para comprender los patrones de tarificación y detectar posibles disparidades o factores de riesgo relevantes en los costos del seguro médico.



Metodología
Para llevar a cabo el análisis, se siguieron los siguientes pasos metodológicos:

Carga y limpieza de datos:
Se utilizó el paquete pandas para importar el conjunto de datos desde un archivo CSV alojado en GitHub. Posteriormente, se eliminaron los valores nulos para asegurar la integridad de los análisis.

Exploración de variables:
Se identificó la variable dependiente como charges (cargos por seguro médico) y se realizó una evaluación preliminar de las variables independientes (age, sex, bmi, children, smoker, region) mediante histogramas para detectar anomalías y evaluar su distribución.

Pruebas de hipótesis y análisis ANOVA:
Se aplicó un análisis de varianza (ANOVA) para determinar si existen diferencias significativas entre los grupos categóricos de variables como sex, smoker y region. Esto permitió contrastar la hipótesis nula de igualdad entre medias poblacionales para las distintas categorías de cada variable.

Pruebas post-hoc (Tukey HSD):
Para aquellas variables donde se encontraron diferencias significativas mediante ANOVA, se aplicó la prueba de comparaciones múltiples de Tukey, con el fin de identificar entre qué grupos específicos existen dichas diferencias.

Visualización de resultados:
Se generaron gráficos de histogramas y comparaciones post-hoc para ilustrar visualmente la distribución de las variables y las diferencias detectadas entre grupos.

MARCO TEORICO
Seguros Médicos y Tarificación
Los seguros médicos son contratos entre un asegurado y una compañía aseguradora que cubren total o parcialmente los gastos médicos derivados de enfermedades, accidentes u otros eventos relacionados con la salud. Uno de los aspectos fundamentales en la gestión de estos seguros es la tarificación, es decir, la determinación de los montos que los clientes deben pagar por sus pólizas. Este cálculo se basa en modelos actuariales y análisis estadísticos que consideran múltiples factores de riesgo.

Variables Predictoras de Costos Médicos
Las aseguradoras utilizan datos demográficos y de estilo de vida para predecir los costos médicos de cada individuo. Entre las variables más relevantes se encuentran:

Edad: Los costos tienden a aumentar con la edad, ya que se incrementa la probabilidad de enfermedades crónicas.

Sexo: Aunque hay variabilidad según el contexto, algunos estudios indican que los hombres tienden a incurrir en mayores costos en ciertas etapas de la vida.

Índice de Masa Corporal (BMI): El sobrepeso y la obesidad están fuertemente asociados con enfermedades como diabetes, hipertensión y problemas cardiovasculares.

Número de hijos: Puede influir en los planes familiares y en los costos generales de aseguramiento.

Tabaquismo: Fumar es uno de los principales factores de riesgo para múltiples enfermedades graves, y suele ser un criterio importante en la tarificación.

Región geográfica: Las diferencias en el costo de vida y la infraestructura médica hacen que el lugar de residencia influya significativamente en los precios.


```python
import pandas as pd
df =pd.read_csv("https://raw.githubusercontent.com/NaomiTapia/Estadisticaverano/refs/heads/main/insurance.csv")
df=df.dropna()
df
```

```python
#Mi variable dependiente (y) es charges
#age:No hay anomalias en el histograma
#sex:Si hubo diferencia significativa entre grupos, male:1 female:2
#bmi:No hay anomalias en el histograma
#children:No hay anomalias en el histograma
#smoker: Si hubo diferencia significativa entre grupos, yes:1 no:2
#region: Si hubo diferencia significativa entre grupos
```

```python
# @title age

from matplotlib import pyplot as plt
df['age'].plot(kind='hist', bins=20, title='age')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

modelo_lineal=ols("charges ~ C(sex)",data=df).fit()
tabla_anova=sm.stats.anova_lm(modelo_lineal)
print(tabla_anova)
#Ho:Todas las medias poblacionales son iguales.
#H1:Al menos una media poblacional es diferente.
#alpha:0.05
#Anova:0.036133 Se rechaza la hipotesis nula, al menos una de ellas es diferencte (si es mas chico se rechaza)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['charges'], groups=df['sex'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="tirmpo")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

```python
# @title children

from matplotlib import pyplot as plt
df['children'].plot(kind='hist', bins=20, title='children')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title bmi

from matplotlib import pyplot as plt
df['bmi'].plot(kind='hist', bins=20, title='bmi')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

modelo_lineal=ols("charges ~ C(smoker)",data=df).fit()
tabla_anova=sm.stats.anova_lm(modelo_lineal)
print(tabla_anova)
#Ho:Todas las medias poblacionales son iguales.
#H1:Al menos una media poblacional es diferente.
#alpha:0.05
#Anova:0.0000 Se rechaza la hipotesis nula, al menos una de ellas es diferencte (si es mas chico se rechaza)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['charges'], groups=df['smoker'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="tirmpo")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

```python
df["smoker"]= df["smoker"].map({"yes":1,"no":2})
df
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

modelo_lineal=ols("charges ~ C(region)",data=df).fit()
tabla_anova=sm.stats.anova_lm(modelo_lineal)
print(tabla_anova)
#Ho:Todas las medias poblacionales son iguales.
#H1:Al menos una media poblacional es diferente.
#alpha:0.05
#Anova:0.030893 Se rechaza la hipotesis nula, al menos una de ellas es diferencte (si es mas chico se rechaza)
```

```python
df["region"].unique()
```

```python
from scipy.stats import levene

# Supón que tienes un DataFrame df con columnas 'respuesta' y 'grupo'
grupo1 = df[df['region'] == 'southwest']['charges']
grupo2 = df[df['region'] == 'southeast']['charges']
grupo3 = df[df['region'] == 'northwest']['charges']
grupo4 = df[df['region'] == 'northeast']['charges']

stat, p = levene(grupo1, grupo2, grupo3, grupo4)
print(f"p-valor de Levene: {p}")
#Ho:Las varianzas de los grupos son iguales. (No hay diferencia significativa en las varianzas entre los grupos comparados).
#H1:Al menos una de las varianzas de los grupos es diferente. (Existe una diferencia significativa en las varianzas entre al menos dos grupos).
#alpha:0.05
#Levene:0.0008610, Se recha la hipotesis nula, no tiene caso hacer BoxCox, pasamos a Tukey para observar de mejor manera
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['charges'], groups=df['region'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="tirmpo")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
#Todas tocan con una linea vertical por lo tanto si pero muy poquito
#Conclusion: Realmente no hay diferencia significativa entre la region y el precio que se paga por los servicios de salud
```
